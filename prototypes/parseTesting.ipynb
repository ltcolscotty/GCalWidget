```python
import requests
import os
from dotenv import load_dotenv
import icalendar
import random
import re

cwd = os.getcwd()
load_dotenv()

url = os.getenv('CANVAS_LINK')
cal_filename = "canvas_cal.ics"

cal_filepath = cwd + f"\\{cal_filename}"
```

```python
try:
    response = requests.get(url)
    response.raise_for_status()

    with open(cal_filename, 'wb') as f:
        f.write(response.content)
    print(f"'{cal_filename}' downloaded successfully!")
    
except requests.exceptions.RequestException as e:
    print(f"An error occurred: {e}")
```

```python
course_list = {}

def parse_ics_file(file_path):
    with open(file_path, 'rb') as icalfile:
        gcal = icalendar.Calendar.from_ical(icalfile.read())
        for component in gcal.walk():
            if component.name == "VEVENT":
                summary = component.get('summary')
                startdt = component.get('dtstart').dt
                url = component.get('url')

                course_list[summary] = [startdt, url]

parse_ics_file(cal_filepath)
```

```python
course_data_regex = r"(?<=\[).*?(?=\])"
assignment_title_regex = r"^([^\[]*)"
course_name_regex = r"[A-Z]{3}[0-9]{0,3}" #made last 3 digits optional (specifically for ORG)
section_regex = r"[0-9]{4}(.*?)[A-Za-z]+" 


#select = random.choice(list(course_list.keys()))
for course in course_list:
    assignment_name = re.match(assignment_title_regex, course).group().strip()
    section_info = re.search(course_data_regex, course).group().strip()
    course_name = re.search(course_name_regex, section_info).group().strip()
    section_name = re.search(section_regex, section_info).group().strip()

    print(course)
    print(assignment_name)
    print(course_name)
    print(section_name)
    
```
